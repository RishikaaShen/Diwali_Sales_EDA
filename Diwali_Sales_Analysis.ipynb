```python
# Exploratory Data Analysis
```

```python
### Gender
```

```python
dataframe.columns
```

```python
# plotting a bar chart for Gender and it's count

axes = sns.countplot(x = 'Gender',data = dataframe, hue = 'Gender')

for bars in axes.containers:
    axes.bar_label(bars)
```

```python
dataframe.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# plotting a bar chart for gender vs total amount

sales_gender = dataframe.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gender, hue = 'Gender')
```

```python
# *From above graphs we can see that most of the buyers are females and the purchasing power of females are greater than men*
```

```python
### Age
```

```python
axes = sns.countplot(data = dataframe, x = 'Age Group', hue = 'Gender')

for bars in axes.containers:
    axes.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = dataframe.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, hue = 'Age Group')
```

```python
#*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*
```

```python
### State
```

```python
# total number of orders from top 10 states

sales_state = dataframe.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders', hue = 'State')
```

```python
# total amount/sales from top 5 states

sales_state = dataframe.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount', hue = 'State')
```

```python
# *From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*
```

```python
### Marital Status
```

```python
axes = sns.countplot(data = dataframe, x = 'Marital_Status', hue = 'Gender')

sns.set(rc={'figure.figsize':(7,5)})
for bars in axes.containers:
    axes.bar_label(bars)
```

```python
sales_state = dataframe.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
# *From above graphs we can see that most of the buyers are married women*
```

```python
### Occupation
```

```python
sns.set(rc={'figure.figsize':(20,5)})
axes = sns.countplot(data = dataframe, x = 'Occupation', hue = 'Occupation')

for bars in axes.containers:
    axes.bar_label(bars)
```

```python
sales_state = dataframe.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount', hue = 'Occupation')
```

```python
# *From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*
```

```python
### Product Category
```

```python
sns.set(rc={'figure.figsize':(25,6)})
axes = sns.countplot(data = dataframe, x = 'Product_Category', hue = 'Product_Category')

for bars in axes.containers:
    axes.bar_label(bars)
```

```python
sales_state = dataframe.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount', hue = 'Product_Category')
```

```python
# *From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*
```

```python
sales_state = dataframe.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders', hue = 'Product_ID')
```

```python
# top 10 most sold products (same thing as above)

fig1, axes1 = plt.subplots(figsize=(12,7))
dataframe.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
